````python
"""
Longform finegrained rewards v2.
- Compared to v1, this version tags the spans for rubrics using LLM judgement.
"""
import logging
import re
from typing import Any, Dict, Optional

from open_instruct.search_rewards.utils.format_utils import extract_answer_context_citations, compute_format_reward
from open_instruct.search_rewards.utils.rubric_utils import _score_property_with_spans
from open_instruct.search_rewards.utils.search_utils import score_num_in_context_search_turns
from open_instruct.search_rewards.utils.citation_utils import score_in_context_citations
from open_instruct.search_rewards.utils.finegrained_utils import (
    FinegrainedScore,
    find_format_reward_spans,
    find_search_turns_reward_spans,
    find_rubric_reward_spans,
    find_citation_reward_spans,
)

LOGGER = logging.getLogger(__name__)


REWARD_WEIGHTS = {
    "rubric_reward": 0.7,
    "citation_reward": 0.1,
    "format_reward": 0.1,
    "num_search_turns_reward": 0.1,
}


def compute_longform_finegrained_reward(response: str, ground_truth: Dict[str, Any], question: str, mcp_parser_name: Optional[str] = None) -> Dict[str, Any]:
    """
    Compute longform finegrained reward with spans.
    
    Returns:
        Dict with:
            - finegrained_scores: List of (score, (start_char, end_char), reward_group_id, response_idx) tuples
            - log_values: Dict of metrics for logging
    """
    extracted_context, extracted_answer, extracted_citations = extract_answer_context_citations(response)
    
    # Initialize result structure
    finegrained_scores = []
    reward_group_id_counter = 0
    reward = 0.0
    
    # score format
    format_reward = compute_format_reward(response, mcp_parser_name=mcp_parser_name)
    format_spans = find_format_reward_spans(response)
    finegrained_scores.append(FinegrainedScore(
        score=format_reward,
        effective_spans=format_spans,
        reward_group_id=reward_group_id_counter,
        reward_group_name="format",
    ))
    reward_group_id_counter += 1
    reward += format_reward * REWARD_WEIGHTS["format_reward"]
    
    
    # score num search turns
    num_search_turns_reward, num_search_turns = score_num_in_context_search_turns(extracted_context, mcp_parser_name=mcp_parser_name)
    num_search_turns_spans = find_search_turns_reward_spans(response, num_search_turns_reward)
    finegrained_scores.append(FinegrainedScore(
        score=num_search_turns_reward,
        effective_spans=num_search_turns_spans,
        reward_group_id=reward_group_id_counter,
        reward_group_name="num_search_turns",
    ))
    reward_group_id_counter += 1
    reward += num_search_turns_reward * REWARD_WEIGHTS["num_search_turns_reward"]
    
    
    # score rubric
    rubric_rewards = []
    for rubric in ground_truth["Answer Critical"]:
        rubric_scores = _score_property_with_spans(extracted_answer, ground_truth, rubric["Ingredient"])
        rubric_reward = sum(rubric_scores.values()) / len(rubric_scores)
        rubric_spans = find_rubric_reward_spans(extracted_answer, rubric_reward)
        finegrained_scores.append(FinegrainedScore(
            score=rubric_reward,
            effective_spans=rubric_spans,
            reward_group_id=reward_group_id_counter,
            reward_group_name=rubric["Ingredient"],
        ))
        rubric_rewards.append(rubric_reward)
        reward_group_id_counter += 1
    reward += sum(rubric_rewards) / len(rubric_rewards) * REWARD_WEIGHTS["rubric_reward"]
    
    
    # score citation
    citation_reward = score_in_context_citations(question, response, extracted_citations)
    citation_spans = find_citation_reward_spans(response, extracted_citations, citation_reward)
    finegrained_scores.append(FinegrainedScore(
        score=citation_reward,
        effective_spans=citation_spans,
        reward_group_id=reward_group_id_counter,
        reward_group_name="citation",
    ))
    reward_group_id_counter += 1
    reward += citation_reward * REWARD_WEIGHTS["citation_reward"]
    
    # Create log values for tracking
    log_values = {
        "format_reward": format_reward,
        "num_search_turns_reward": num_search_turns_reward,
        "rubric_reward": sum(rubric_rewards) / len(rubric_rewards),
        "citation_reward": citation_reward,
        "overall_reward": reward,
        "response_length": len(response),
        "answer_extracted": extracted_answer is not None,
        "num_citations": len(extracted_citations) if extracted_citations else 0,
        "num_search_turns": num_search_turns,
    }
    
    return {
        "finegrained_scores": finegrained_scores,
        "log_values": log_values,
    }

````

````python
import logging
import os
from typing import Any, Dict, List
from openai import OpenAI
from pydantic import BaseModel

from open_instruct.search_rewards.utils.run_utils import extract_json_from_response, run_litellm


client = OpenAI()


class SingleCriterionOutput(BaseModel):
    criterion: str
    score: float
    spans: List[str]
    reason: str
    
class LongRubricOutput(BaseModel):
    scores: List[SingleCriterionOutput]
    overal_score: float

LONG_REASONING_RUBRIC = """
You are an expert in evaluating text quality. You will receive a question, a knowledge collection process, a response, and some rubrics to help you evaluate the response.
The input will be given in the following format:
- question: an open-ended question that needs to be answered with knowledge source attribution
- background: a search process where an agent asks a series of search queries and notes the corresponding search results
- response: a final response to the question with citations to the knowledge source
- rubrics: a list of rubrics to facilitate evaluating the response

Your task is to evaluate the response based on the above information. You need to go over the below aspects and provide your evaluation in a detailed json format. You need to provide the score for each creterion, e.g., rubric, citation claims, etc. in a fine-grained manner. Return a score on a scale of 0 to 2 indicating how appropriate the response is based on each criterion. You should output the spans in the format of verbatim snippets from the response that are judged as satisfying the criterion. Output JSON in the format: {{"criterion": x, "score": y, "spans": [z], "reason": r}}.

**Rubric scores** Output the score for every rubric in the provided rubric list; 

**Citation scores** Evaluate whether each citation can accurately support the wrapped claims; whether the claims are well covered by citations, if not, penalize those claims that could have cited one of the evidence in context.

**Format scores** Check whether the final answer is put between <answer></answer> tags; whether each claim is using a valid snippet id that can be found in context.

**Search scores** Evaluate the overall search quality and the quality of each search query; the search should be exploratory and strategic in order to collect all necessary knowledge and revise the plan based on observations of the search results.

Now start your evaluation. Remember to provide a detailed finegrained evaluation: Output JSON in the format: {{"criterion": x, "score": y, "spans": [z], "reason": r}}.
"""

LONG_REASONING_RUBRIC = """
# Long Reasoning Evaluation Rubric

You are an expert evaluator of text quality. You will assess a response based on provided inputs and return detailed JSON scores.

## Input Format
- **question**: Open-ended question requiring knowledge source attribution
- **background**: Agent's search process with queries and results
- **response**: Final answer with citations to knowledge sources
- **rubrics**: Evaluation criteria list

## Task
Evaluate the response across four key aspects using a 0-2 scale for each criterion. Return detailed JSON for each evaluation.

## Evaluation Aspects

### 1. Rubric Scores
Score the response against each provided rubric criterion.

### 2. Citation Scores
- Do citations accurately support their wrapped claims?
- Are all supportable claims properly cited?
- Penalize uncited claims that could reference available evidence.
- Note that when a claim is cited but irrelevant to the question, you should still give a score of 0 because the model might be trying to hack the citation score.
- Try to give a score based on the quality of the claim, as well as the coherence of the claim to the entire response.

### 3. Format Scores
- Is the final answer enclosed in `<answer></answer>` tags?
- Do all snippet IDs exist in the provided context?

### 4. Search Scores
- Overall search strategy quality
- Individual query effectiveness
- Search exploration and adaptability based on results

## Output Format
For each criterion, return JSON:
```json
[
{
  "criterion": "rubric 1 from the input rubrics",
  "score": 0-2,
  "spans": ["verbatim snippets from response"],
  "reason": "detailed explanation"
},
{
  "criterion": "rubric 2 from the input rubrics",
  "score": 0-2,
  "spans": ["verbatim snippets from response"],
  "reason": "detailed explanation"
},
...
{
  "criterion": "citation for claim x",
  "score": 0-2,
  "spans": ["verbatim snippets from response"],
  "reason": "detailed explanation"
}
...
]
```

Note that you need to provide scores for all rubrics separately, as well as for each citation claim. Provide fine-grained evaluation with specific evidence from the response.
Additionally, you need to provide an overall score for the response based on the fine-grained rubric scores and the citation scores. Weight higher for the 
"""

LOGGER = logging.getLogger(__name__)
def _score_long_rubrics_with_spans_and_reason(response: str, question: str, background: str, rubrics: str) -> float:
    system_prompt = LONG_REASONING_RUBRIC
    user_prompt = (
        f"""<question>{question}</question>\n<background>{background}</background>\n<response>{response}</response>\n<rubrics>{rubrics}</rubrics>"""
    )

    # wrap in try-except to handle litellm API errors
    # these might just be ephemeral, so we don't want to crash the whole training job.
    try:
        resp = client.responses.parse(
            input=user_prompt,
            model=os.environ.get("RUBRIC_JUDGE_MODEL", "gpt-5-nano"),
            text_format=LongRubricOutput,
        )
        return resp
            
    except Exception as e:
        LOGGER.warning(f"Error scoring rubric: {e}")
        return LongRubricOutput(scores=[SingleCriterionOutput(criterion="", score=0.0, spans=[], reason="")])
````
